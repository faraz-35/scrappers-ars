```python
import pymysql.cursors
import Classifier 
import the_national_interest
mydb = pymysql.connect(
  host="localhost",
  user="root",
  password="abc",
  database="nexus360",
)

mycursor = mydb.cursor()
data=the_national_interest.run_scrapper()
data1=data
data1["Category"]=""
for i in range(len(data)):
    data1["Category"][i]=Classifier.classify_articles(data1["Article"][i])


sql = "INSERT INTO nexus_360_news_post (title, News_content,date,category,author,dateOfExtraction,link,source) VALUES (%s, %s,%s, %s,%s,%s,%s, %s)"
val = (data1["Title"][1], data1["Article"][0],data1["DateOfPublication"][0],data1["Category"][0],1,data1["DateOfExtraction"][0],data1["Link"][0],data1["Source"][0])
mycursor.execute(sql, val)

mydb.commit()

print(mycursor.rowcount, "record inserted.")
```
